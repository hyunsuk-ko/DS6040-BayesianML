```python
import pandas as pd
import numpy as np
from scipy.stats import beta
import matplotlib.pyplot as plt
```

```python
X = 25
N = 50
mean = [0.25, 0.5, 0.75, 0.9]
sample_size = [2, 10, 50, 100]
```

```python
params = []
for m in mean:
    for n in sample_size:
        a = n * m
        b = round(n - a,3)
        params.append((a,b))
```

```python
a, b = 0.5, 1.5
x = np.linspace(beta.ppf(0.01, a, b),beta.ppf(0.99, a, b), 100)
fig, axs = plt.subplots(4, 4, constrained_layout=True, figsize = (8,5))

count = 0
for i in range (0,4):
    for j in range(0,4):
        axs[i, j].plot(x, beta.pdf(x, params[count][0] + X, params[count][1] + N - X), 'r-')
        axs[i, j].set_title(f'Beta(a={params[count][0]}, b={params[count][1]})')
        count += 1
        
plt.suptitle(f"Beta Distributions (X = {X}, N = {N})", fontsize=14)
```

```python
X = 50
N = 100

fig, axs = plt.subplots(4, 4, constrained_layout=True, figsize = (8,5))

count = 0
for i in range (0,4):
    for j in range(0,4):
        axs[i, j].plot(x, beta.pdf(x, params[count][0] + X, params[count][1] + N - X), 'r-')
        axs[i, j].set_title(f'Beta(a={params[count][0]}, b={params[count][1]})')
        count += 1
        
plt.suptitle(f"Beta Distributions (X = {X}, N = {N})", fontsize=14)
```

The highest peak is located at expected value: E(X) = mean = $\frac{a}{a+b}$, and as the number of observation increases, the width of distribtuion gets narrower.
