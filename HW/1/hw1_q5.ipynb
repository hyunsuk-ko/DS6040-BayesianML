```python
cont_columns = ['sbp', 'tobacco', 'ldl', 'adiposity', 'typea', 'obesity', 'alcohol', 'age']
```

```python
for column in cont_columns:
    X = chd[[column]]
    mu = chd[[column]].mean()
    sd = chd[[column]].std()
    chd[column] = (X - mu) / sd
chd.head()
```

```python
chd_no = chd[chd['chd'] == 0]
chd_yes = chd[chd['chd'] == 1]
chd_no
```
